### Library importing
<br>

Most of the Matplotlib utilities lie under the pyplot submodule, and are usually imported under the __plt__ alias:

```python
import matplotlib.pyplot as plt
```

We will create two waveforms using NumPy library and plot them using Matplotlib library

```python
import numpy as np 
x = np.linspace(0, 15, 200)

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

#### Labeling plots
We add axes titles, legend and figure title

```python
plt.plot(x, np.sin(x), '-',label='Sine of X')
plt.plot(x, np.cos(x), '--',label='Cosine of X' )

plt.legend(loc='lower right')
plt.xlabel('Time (Xronos)')
plt.ylabel('Amplitude (Platos)')
plt.title('Grafima Simaton')
```

#### Customise Plot
There are many attributes we can use to customise the plot:
- `linestyle` or `ls` using 'dashdot'  or  '-.',   'dotted' or ':'
- `color` using 'red', 'r', RGB tuple(1.0,0.2,0.3) 
- `marker` using "*", "o" ,"D" for diamonds
- combination of the above ':c'for dotted cyan

```python
plt.plot(x, np.cos(x), ls='-.',color = 'r');
```

#### Adjust the axes limits

```python
plt.plot(x, np.cos(x), ':c');
plt.xlim(3, 12.2)
plt.ylim(-2.5, 2.5);
```

#### Multiple plots on the same figure:Subplots
The subplot method takes three arguments to define the figure window: the number of rows (2), the number of columns (1) and the plot number (1).

```python
plt.subplot(211)
plt.plot(x, np.sin(x), '-')
plt.subplot(212)
plt.plot(x, np.cos(x), '--')
```

adding labels

```python
plt.subplot(211)
plt.plot(x, np.sin(x), '*')
plt.ylabel('Amplitude (Platos)')
plt.title('Grafima Sine of X')
plt.subplot(212)
plt.plot(x, np.cos(x), 'D')
plt.xlabel('Time (Xronos)')
plt.ylabel('Amplitude (Platos)')
plt.title('Grafima Cosine of X')

plt.subplots_adjust(wspace=0.4,hspace=0.7) #set the distances between subplots
```

# Plots for categorical features


## Box Plot

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers.
<br>

`median (Q2/50th Percentile)`: the middle value of the dataset.

`first quartile (Q1/25th Percentile)`: is the median of the data points below of the dataset median

`third quartile (Q3/75th Percentile)`: is the median of the data points above of the dataset median

`interquartile range (IQR)`: 25th to the 75th percentile.

`whiskers` display  the thresholds beyond which values are considered as outliers

`upper threshold`: Q3 + 1.5*IQR

`lower threshold`: Q1 -1.5*IQR

```python
sns.boxplot(x=data.origin, y=data.mpg)
plt.axhline(data.mpg.mean(),color='r',linestyle='dashed',linewidth=2)
```

```python
sns.boxplot('cylinders', 'mpg',data=data)
```

#### Seaborn violin plot
Similar to box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution

```python
sns.violinplot('origin', 'mpg',data=data,palette="Set2")
```

```python
sns.violinplot('cylinders', 'mpg',data=data,palette="Set2")
```
